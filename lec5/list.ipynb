#  Editing a List

We can **add**, **remove**, or **count** items in a list.

```python
shopping_list = ["apple", "banana", "milk"]

shopping_list.append("eggs")     # Add new item
shopping_list.remove("banana")   # Remove item
print(len(shopping_list))        # Count items

print(shopping_list)


````python
shopping_list.append("egges")
shopping_list.remove("egges")
shopping_list.extend([1,3,True,25.5])
print(len(shopping_list))
````

# Using Lists in Robotics

We can use lists to store **movement commands** for a robot car!

```python
directions = ["forward", "left", "forward", "right", "stop"]

for move in directions:
    print("Robot is moving:", move)


````python
# A list of robot car directions
directions = ["forward", "left", "forward", "right", "stop"]

# The robot follows each direction
for move in directions:
    print("Robot is moving:", move)
````

#  Lists â€” Main Concepts

1.  **List items are enclosed in square brackets**  
   Example: [1, 2, 3, 4]

2.  **Lists are ordered**, so we can use the index to access any item.  
   Example: numbers[0] â†’ first item.

3.  **Lists are mutable**, meaning we can add, delete, or edit elements.

4.  **List items are not unique** â€” the same value can appear more than once.

5.  **Lists can have different data types** â€” numbers, strings, booleans, etc.
